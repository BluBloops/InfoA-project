```python
import math
import time
from algoviz.svg import SVGView, Text
from dungeon import Dungeon
from player import Player
from enemy import Enemy
from item import Item

def fight(ens, p1, A):
    temp = 0

    for eZeiger in ens:
        if math.sqrt((eZeiger.get_x() - p1.get_x()) ** 2 + (eZeiger.get_y() - p1.get_y()) ** 2) < (eZeiger.get_width() + p1.get_radius()):
            t = Text("FIGHT!", 230, 300, A)
            t.set_attribute("font-size", 50)
            t.set_attribute("fill", "#FF0000")

            eZeiger.set_e_health(eZeiger.get_e_health() - p1.get_dmg())
            temp += eZeiger.get_dmg()
            p1.set_color("red")
            time.sleep(1)
            t.__del__()

    return temp

def enemy_move(ens, A, D):
    for enemy in ens:
        enemy.enemy_move(A, D)

first_dungeon = Dungeon()
first_dungeon.spawn_dungeon()

# List of Enemies
enemies = [Enemy(1000, 150) for _ in range(10)]
for enemy in enemies:
    enemy.spawn_enemy(first_dungeon.dungeon, first_dungeon)

# List of Items
items = [Item() for _ in range(20)]
for item in items:
    item.spawn_item(first_dungeon.dungeon, first_dungeon)

player01 = Player(1000, 5, 400, True)
player01.spawn_player(first_dungeon.dungeon, first_dungeon)

while player01.get_p_health() > 0 and enemies:
    player01.player_move(first_dungeon.dungeon, first_dungeon)
    enemy_move(enemies, first_dungeon.dungeon, first_dungeon)

    if player01.get_damage_enable():
        player01.set_p_health(player01.get_p_health() - fight(enemies, player01, first_dungeon.dungeon))
    else:
        fight(enemies, player01, first_dungeon.dungeon)

    # Item Collision
    for item in items:
        if item.get_x() == player01.get_x() and item.get_y() == player01.get_y():
            item.move_to(0, 0)
            item.remove_from_view()
            kind = item.get_kind()
            if kind == 0:
                # Speed
                player01.inc_p_speed(5)
            elif kind == 1:
                # Health
                player01.set_p_health(player01.get_p_health() + 200)
            elif kind == 2:
                # Invincibility
                player01.set_color("yellow")
                player01.set_damage_enable(False)

Text("GAME OVER!", 140, 300, first_dungeon.dungeon).set_attribute("font-size", 50).set_attribute("fill", "#FFFFFF").set_attribute("stroke", "#000000")
```
