```python
import math
import time
from algoviz.svg import SVGView, Text
from dungeon import Dungeon
from player import Player
from enemy import Enemy
from item import Item

def fight(enemies, player, svg_view):
    temp = 0

    for eZeiger in enemies:
        if math.sqrt((eZeiger.get_x() - player.get_x()) ** 2 + (eZeiger.get_y() - player.get_y()) ** 2) < (eZeiger.get_width() + player.get_radius()):
            t = Text("FIGHT!", 230, 300, svg_view)
            t.set_attribute("font-size", 50)
            t.set_attribute("fill", "#FF0000")

            eZeiger.set_e_health(eZeiger.get_e_health() - player.get_dmg())
            temp += eZeiger.get_dmg()
            player.set_color("red")
            time.sleep(1)
            t.__del__()

    return temp

def enemy_move(enemies, svg_view, dungeon):
    for enemy in enemies:
        enemy.enemy_move(svg_view, dungeon)

first_dungeon = Dungeon()
first_dungeon.spawn_dungeon()

# List of Enemies
enemies = [Enemy(1000, 150) for _ in range(10)]
for enemy in enemies:
    enemy.spawn_enemy(first_dungeon.get_dungeon(), first_dungeon)

# List of Items
items = [Item() for _ in range(20)]
for item in items:
    item.spawn_item(first_dungeon.get_dungeon(), first_dungeon)

player01 = Player(1000, 5, 400, True)
player01.spawn_player(first_dungeon.get_dungeon(), first_dungeon)

while player01.get_p_health() > 0 and enemies:
    player01.player_move(first_dungeon.get_dungeon(), first_dungeon)
    enemy_move(enemies, first_dungeon.get_dungeon(), first_dungeon)

    if player01.get_damage_enable():
        player01.set_p_health(player01.get_p_health() - fight(enemies, player01, first_dungeon.get_dungeon()))
    else:
        fight(enemies, player01, first_dungeon.get_dungeon())

    # Item Collision
    for item in items:
        if item.get_x() == player01.get_x() and item.get_y() == player01.get_y():
            item.move_to(0, 0)
            item.remove_from_view()
            kind = item.get_kind()
            if kind == 0:
                # Speed
                player01.inc_p_speed(5)
            elif kind == 1:
                # Health
                player01.set_p_health(player01.get_p_health() + 200)
            elif kind == 2:
                # Invincibility
                player01.set_color("yellow")
                player01.set_damage_enable(False)

game_over_text = Text(140, 300, "GAME OVER!", first_dungeon.get_dungeon())
game_over_text.set_attribute("font-size", 50)
game_over_text.set_attribute("fill", "FFFFFF")
game_over_text.set_attribute("stroke", "000000")
```
